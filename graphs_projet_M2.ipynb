```python
import matplotlib.pyplot as plt

# Données
mdf_base = 99.938582
mdf_llama = [mdf_llama1 := 4.136894, mdf_llama2 := 2.693395, mdf_llama4 := 1.455700]
mdf_gpt = [mdf_gpt1 := 3.899416, mdf_gpt2 := 2.461489, mdf_gpt4 := 2.279228]

# Calcul de l'accélération
acceleration_llama = [mdf_base / time for time in mdf_llama]
acceleration_gpt = [mdf_base / time for time in mdf_gpt]

# Création du graphe
plt.figure(figsize=(10, 6))
plt.plot(['1 Noeud', '2 Noeuds', '4 Noeuds'], acceleration_llama, marker='o', label='Llama')
plt.plot(['1 Noeud', '2 Noeuds', '4 Noeuds'], acceleration_gpt, marker='o', label='GPT')
plt.xlabel('Nombre de Noeuds')
plt.ylabel('Accélération')
plt.title('Accélération en MPI')
plt.legend()
plt.grid(True)
plt.show()
```
